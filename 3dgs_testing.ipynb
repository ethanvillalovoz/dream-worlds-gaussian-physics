```python
import pycolmap

def undistort_files(base_path: str, images_path: str):
    pycolmap.undistort_images(
        output_path=f"./{base_path}/undistorted/", 
        input_path=f'{base_path}/sparse/0', 
        image_path=f'{base_path}/{images_path}'
    )

undistort_files('data/rocinante', 'train')
```

```python
import gaussian_splatting

gaussian_splatting.train -s data/rocinante/undistorted -d output/roci -i 15000 --mode densify
```
